### 1.3 Exceptions

Write `safe_divide(a, b)` that:
- Returns `a / b` normally
- Returns `None` if `b == 0` (catch `ZeroDivisionError`)
- Raises `TypeError` with message `"Inputs must be numeric"` if either input is not a number

````python
def safe_divide(a, b):
    if not isinstance(a, (int, float)) or not isinstance(b, (int, float)):
        raise TypeError("Inputs must be numeric")

    try:
        return a / b
    except ZeroDivisionError:
        return None

# Tests
assert safe_divide(10, 2) == 5.0
assert safe_divide(5, 0) is None

try:
    safe_divide('x', 2)
    assert False, "Should have raised TypeError"
except TypeError as e:
    assert str(e) == "Inputs must be numeric"

print("1.3 passed ✓")
````

### 4.3 Eigenvalues & Eigenvectors

For the matrix:
```
A = [[4, 1],
     [2, 3]]
```

1. Compute eigenvalues and eigenvectors using `np.linalg.eig`.
2. **Verify** the eigenvector equation: `A @ v = λ * v` for each eigenpair.
3. Plot both eigenvectors as arrows, scaled by their eigenvalues, on a 2D plot.
4. **Explain** in a markdown cell (below): what does it geometrically mean for a matrix to stretch a vector?

````python
import numpy as np
import matplotlib.pyplot as plt

A = np.array([[4, 1], [2, 3]], dtype=float)

# Compute eigenvalues and eigenvectors
eigenvalues, eigenvectors = np.linalg.eig(A)

print(f"Eigenvalues: {eigenvalues}")
print(f"Eigenvectors (columns): \n{eigenvectors}")

# Verify Av = λv for each eigenpair
for i in range(len(eigenvalues)):
    v = eigenvectors[:, i]
    lam = eigenvalues[i]

    Av = A @ v
    lv = lam * v

    print(f"\nEigenpair {i+1}:")
    print("A @ v =", Av)
    print("λ * v =", lv)

    assert np.allclose(Av, lv), "Eigenvector equation failed!"

# Plot eigenvectors scaled by eigenvalues
fig, ax = plt.subplots()

for i in range(len(eigenvalues)):
    v = eigenvectors[:, i]
    scaled_v = eigenvalues[i] * v

    ax.quiver(0, 0, scaled_v[0], scaled_v[1],
              angles='xy', scale_units='xy', scale=1,
              label=f'λ{i+1}v')

ax.set_xlim(-5, 5)
ax.set_ylim(-5, 5)
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_title("Eigenvectors Scaled by Eigenvalues")
ax.axhline(0, color='k', lw=0.8)
ax.axvline(0, color='k', lw=0.8)
ax.legend()

plt.grid(True)
plt.tight_layout()
plt.show()

print("4.3 passed ✓")
````

**Your Geometric Explanation (4.3):**

> *Replace this text with your explanation of what eigenvalues/eigenvectors mean geometrically — what does it mean for a matrix to "stretch" or "squish" a vector?*

### Geometric Explanation

Geometrically, eigenvectors are special vectors that keep the same direction when a matrix transformation is applied. The matrix may stretch or shrink these vectors, but it does not rotate them.

The eigenvalue tells us how much the vector changes in length:
- If the eigenvalue is greater than 1, the vector is stretched.
- If it is between 0 and 1, the vector is shrunk.
- If it is negative, the vector flips direction.

This means a matrix "stretching" or "squishing" a vector describes how the transformation changes its magnitude while preserving its direction for eigenvectors.

**SVD → PCA Connection (4.4):**

> *Explain: what are U, S, Vt in SVD, and how do they relate to principal components? Which matrix contains the directions of maximum variance?*

### SVD and PCA Relationship

Singular Value Decomposition (SVD) breaks a matrix into three components: U, S, and Vᵀ.

- **U** contains the left singular vectors, which represent directions in the transformed output space.
- **S** contains singular values, which indicate the importance or strength of each direction.
- **Vᵀ** contains the right singular vectors, which represent principal directions in the original data space.

SVD is closely related to PCA because PCA finds the directions of maximum variance in data. These directions are the principal components, which correspond to the right singular vectors in Vᵀ.

Dimensionality reduction is performed by keeping only the largest singular values and their associated vectors. This preserves the most important structure of the data while reducing complexity

### 5.3 Error Metrics

Given these predictions and actuals, **implement from scratch** (no sklearn):
- MAE, MSE, RMSE
- R² Score
- Adjusted R² (assume 2 features)

````python
import numpy as np

y_true = np.array([3.0, 5.0, 2.5, 7.0, 4.5, 6.0, 1.5, 8.0])
y_pred = np.array([2.8, 5.2, 2.1, 7.5, 4.0, 6.3, 2.0, 7.8])

n, p = len(y_true), 2

# Errors
errors = y_true - y_pred

# MAE
mae = np.mean(np.abs(errors))

# MSE
mse = np.mean(errors ** 2)

# RMSE
rmse = np.sqrt(mse)

# R2 Score
ss_res = np.sum(errors ** 2)
ss_tot = np.sum((y_true - np.mean(y_true)) ** 2)
r2 = 1 - (ss_res / ss_tot)

# Adjusted R2
adj_r2 = 1 - (1 - r2) * (n - 1) / (n - p - 1)

print(f"MAE: {mae:.4f}")
print(f"MSE: {mse:.4f}")
print(f"RMSE: {rmse:.4f}")
print(f"R2: {r2:.4f}")
print(f"Adj. R2: {adj_r2:.4f}")
````

### 5.4 Distribution Testing & Stationarity

1. Generate two samples: `s1` ~ Normal(0, 1), `s2` ~ Exponential(1), both n=200 (seed=0).
2. Run the **Kolmogorov-Smirnov test** (`stats.kstest`) comparing each against a standard normal CDF.
3. Generate a non-stationary time series (linear trend + noise). Run the **Augmented Dickey-Fuller test** (`adfuller` from `statsmodels`) and interpret the result.
4. Difference the series once and re-run ADF to confirm stationarity.

````python
import numpy as np
from scipy import stats
from statsmodels.tsa.stattools import adfuller

np.random.seed(0)

# 1. Generate samples
s1 = np.random.normal(0, 1, 200)
s2 = np.random.exponential(1, 200)

# 2. KS test (compare against N(0,1))
ks_s1 = stats.kstest(s1, 'norm')
ks_s2 = stats.kstest(s2, 'norm')

print(f"KS test s1 (normal): stat={ks_s1.statistic:.4f}, p={ks_s1.pvalue:.4f}")
print(f"KS test s2 (exponential): stat={ks_s2.statistic:.4f}, p={ks_s2.pvalue:.4f}")

# 3. Non-stationary time series
t = np.arange(200)
ts = 0.05 * t + np.random.normal(0, 1, 200)

adf_result = adfuller(ts)

print(f"\nADF p-value: {adf_result[1]:.4f}")
print(f"Stationary? {adf_result[1] < 0.05}")
print(f"ADF statistic: {adf_result[0]:.4f}")

# 4. Differencing once
ts_diff = np.diff(ts)

adf_diff = adfuller(ts_diff)

print(f"\nAfter differencing - p-value: {adf_diff[1]:.4f}, Stationary? {adf_diff[1] < 0.05}")
````

### 5.5 Model Monitoring Concepts

1. Implement `compute_psi(expected, actual, bins=10)` that computes the **Population Stability Index** between two distributions.
2. Simulate a scenario where a model's input distribution shifts and plot both distributions.
3. In the markdown cell, define: *concept drift*, *covariate drift*, PSI thresholds (what values indicate no/minor/major shift), and one trigger for retraining.

````python
def compute_psi(expected, actual, bins=10):
    """
    PSI = sum((actual_pct - expected_pct) * ln(actual_pct / expected_pct))
    Add epsilon=1e-10 to avoid log(0).
    """
    import numpy as np

def compute_psi(expected, actual, bins=10):
    epsilon = 1e-10

    # Create одинаков bins using expected distribution
    breaks = np.linspace(np.min(expected), np.max(expected), bins + 1)

    expected_counts, _ = np.histogram(expected, bins=breaks)
    actual_counts, _ = np.histogram(actual, bins=breaks)

    expected_pct = expected_counts / len(expected)
    actual_pct = actual_counts / len(actual)

    # avoid zero division
    expected_pct = np.where(expected_pct == 0, epsilon, expected_pct)
    actual_pct = np.where(actual_pct == 0, epsilon, actual_pct)

    psi = np.sum((actual_pct - expected_pct) * np.log(actual_pct / expected_pct))
    return psi

np.random.seed(1)
train_dist = np.random.normal(50, 10, 1000)    # training distribution
drift_dist = np.random.normal(65, 12, 1000)    # shifted production distribution

psi_value = compute_psi(train_dist, drift_dist)
print(f"PSI: {psi_value:.4f}")
print(f"Shift severity: {'Major' if psi_value > 0.2 else 'Minor' if psi_value > 0.1 else 'Stable'}")

ax.hist(train_dist, bins=30, alpha=0.6, label="Train Distribution")
ax.hist(drift_dist, bins=30, alpha=0.6, label="Drift Distribution")

ax.legend()
ax.set_title("Distribution Shift Visualization")
````

Concept Drift:
When relationship between input (X) and output (Y) changes over time.

Covariate Drift:
When input distribution changes but relationship between X and Y stays same.

PSI Thresholds:

< 0.1 → Stable
0.1 – 0.2 → Minor drift

0.2 → Major drift

Retraining Trigger:
When PSI > 0.2 or model accuracy drops in production.

````python
total = 10
red, blue, green = 4, 3, 3

# 1. Probabilities
p_red = red / total
p_blue = blue / total
p_green = green / total

# 2. Joint probability: red first, then blue
p_red_then_blue = (red / total) * (blue / (total - 1))

# 3. Conditional probability: blue given red already drawn
p_blue_given_red = blue / (total - 1)

# 4. Independence check
independent = abs(p_blue_given_red - p_blue) < 1e-9

print(f"P(red)={p_red:.3f}, P(blue)={p_blue:.3f}, P(green)={p_green:.3f}")
print(f"P(red,blue): {p_red_then_blue:.4f}")
print(f"P(blue|red): {p_blue_given_red:.4f}")
print(f"P(blue): {p_blue:.4f}")
print(f"Independent? {independent}")
````

### 6.3 Bayes' Theorem

**Scenario — Spam filter:**
- P(Spam) = 0.30 (prior)
- P("free" | Spam) = 0.80 (likelihood)
- P("free" | Not Spam) = 0.05

1. Compute P(Spam | "free") using Bayes' theorem.
2. Implement a simple `naive_bayes_predict(prior_spam, p_word_given_spam, p_word_given_ham)` that returns the posterior P(Spam | word).
3. In the markdown cell, map each term — **prior, likelihood, evidence, posterior** — to the spam example.

````python
p_spam = 0.30
p_ham = 1 - p_spam

p_free_given_spam = 0.80
p_free_given_ham = 0.05

# P(free) using law of total probability
p_free = (p_free_given_spam * p_spam) + (p_free_given_ham * p_ham)

# Bayes theorem
p_spam_given_free = (p_free_given_spam * p_spam) / p_free

print(f"P('free'): {p_free:.4f}")
print(f"P(Spam | 'free'): {p_spam_given_free:.4f}")

def naive_bayes_predict(prior_spam, p_word_given_spam, p_word_given_ham):
    prior_ham = 1 - prior_spam

    p_word = (p_word_given_spam * prior_spam) + (p_word_given_ham * prior_ham)

    posterior = (p_word_given_spam * prior_spam) / p_word
    return posterior

pred = naive_bayes_predict(0.30, 0.80, 0.05)

assert abs(pred - p_spam_given_free) < 1e-9

print(f"naive_bayes_predict: {pred:.4f}")
print("6.3 passed ✓")
````

**Bayes Term Mapping (6.3):**

| Term | Formula | In this example |
|------|---------|-----------------|
| Prior | P(Spam) | The initial probability that an incoming email is spam (0.30). |
| Likelihood | P(word \| Spam) | The probability of seeing the word "free" given that the email is spam (0.80). |
| Evidence | P(word) | The overall probability of seeing the word "free" in any email, regardless of whether it's spam or not (calculated as 0.2750). |
| Posterior | P(Spam \| word) | The updated probability that an email is spam, given that it contains the word "free" (0.8727). |

### 6.4 Central Limit Theorem

1. Define a population following an **exponential distribution** (λ=1, size=100,000).
2. Repeatedly draw samples of size `n=30` and record their means — do this 5,000 times.
3. Plot the histogram of sample means.
4. Overlay the **theoretical normal distribution** predicted by the CLT: μ = population mean, σ = population_std / sqrt(n).
5. Run a KS test to confirm the distribution of sample means is approximately normal.

````python
np.random.seed(7)
population = np.random.exponential(scale=1.0, size=100_000)

n_samples = 5000
sample_size = 30

# 1. Sample means
sample_means = np.array([
    np.mean(np.random.choice(population, sample_size, replace=True))
    for _ in range(n_samples)
])

pop_mean = population.mean()
pop_std = population.std()

# CLT predicted std of sample means
clt_std = pop_std / np.sqrt(sample_size)

print(f"Population mean: {pop_mean:.4f}, Population std: {pop_std:.4f}")
print(f"Sample means mean: {sample_means.mean():.4f}, std: {sample_means.std():.4f}")
print(f"CLT predicted std: {clt_std:.4f}")

# -------------------------
# Plot
# -------------------------
fig, ax = plt.subplots()

ax.hist(sample_means, bins=40, density=True, alpha=0.6)

x = np.linspace(min(sample_means), max(sample_means), 200)
ax.plot(x, stats.norm.pdf(x, pop_mean, clt_std), 'r')

ax.set_title('CLT: Distribution of Sample Means')
plt.tight_layout()
plt.show()

# -------------------------
# KS test
# -------------------------
ks_result = stats.kstest(sample_means, 'norm', args=(pop_mean, clt_std))
print(f"KS test p-value: {ks_result.pvalue:.4f} > Approximately normal? {ks_result.pvalue > 0.05}")
````

**CLT Reflection (6.4):**

> *In 2–3 sentences: Why does this result matter for statistics and machine learning? What assumption does the CLT justify in many classical statistical tests?*

The Central Limit Theorem (CLT) matters in statistics and machine learning because it ensures that the distribution of sample means becomes approximately normal as the sample size increases, even if the original data is not normally distributed. This makes it possible to perform reliable inference and estimation using normal-based methods.

The CLT justifies the assumption of normality of the sampling distribution, which is the foundation of many classical statistical tests such as t-tests, z-tests, ANOVA, and confidence intervals.
